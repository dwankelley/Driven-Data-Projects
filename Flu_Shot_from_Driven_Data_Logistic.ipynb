```python
#remove NAN by inputing the avg value
def rmv_NAN(df_rmv_NAN):
  """ this function removes NAN and replaces it with the average value, column wise"""
  for col in list(df_rmv_NAN):
    avg=df_rmv_NAN[col].mean(skipna=True)
    df_rmv_NAN[col].fillna(avg, inplace=True) 

rmv_NAN(df)
```
